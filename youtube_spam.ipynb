Exploratory Data Analysis (EDA)

 Distribution of CLASS

```python
sns.countplot(x='CLASS',data=df)
plt.title('Distribution of Class')
plt.show()
```

Unique Video Names and Most Commented Videos

```python
print(f"Unique Video Names: {df['VIDEO_NAME'].nunique()}")
most_commented_videos = df['VIDEO_NAME'].value_counts().head(10)
print(most_commented_videos)
```

```python
plt.figure(figsize=(10,8))
sns.barplot(x=most_commented_videos,y=most_commented_videos.index,palette='viridis')
plt.title('Top 10 Most Commented Videos')
plt.xlabel('Numberof Comments')
plt.ylabel('Videos')
plt.show()
```

```python
import plotly.express as px
```

```python
fig_videos = px.bar(
    x=most_commented_videos.values,
    y=most_commented_videos.index,
    labels={'x':'Number of Comments', 'y':'Video Name'},
    title='Top 10 Most Commented Videos',
    orientation='h',
    color=most_commented_videos.values,
    color_continuous_scale='Viridis'
)
fig_videos.update_layout(coloraxis_showscale=False)
fig_videos.show()
```

Unique Authors and Top Commenters

```python
print(f"Unique Authors:{df['AUTHOR'].nunique()}")
```

```python
top_comments=df['AUTHOR'].value_counts().head(10)
print(top_comments)
```

```python
plt.figure(figsize=(10,8))
sns.barplot(x=top_comments.values,y=top_comments.index,palette='coolwarm')
plt.title("Top 10 Comments")
plt.xlabel("Number of Comments")
plt.ylabel('author')
plt.show()
```

```python
fig_authors = px.bar(
    x=top_comments.values,
    y=top_comments.index,
    labels={'x':'Number of Comments', 'y':'Author'},
    title='Top 10 Commenters',
    orientation='h',
    color=top_comments.values,
    color_continuous_scale='Portland'
)
fig_authors.update_layout(coloraxis_showscale=False)
fig_authors.show()
```

Length of Comments

```python
plt.figure(figsize=(8,8))
sns.histplot(df['CONTENT'].apply(len),bins=50)
plt.title('Distribution of comment Length')
plt.show()
```

Word Cloud for Content

```python
from wordcloud import WordCloud
```

```python
word_could=WordCloud(width=800,height=400,background_color='pink').generate(' '.join(df['CONTENT']))
plt.figure(figsize=(10,6))
plt.imshow(word_could,interpolation='bilinear')
plt.axis('off')
plt.show()
```

Temporal Analysis

```python
df['DATE']=pd.to_datetime(df['DATE'])
df.set_index('DATE',inplace=True)
df['CONTENT'].resample('M').count().plot()
plt.title('Number of Comments over time')
plt.show()
```
